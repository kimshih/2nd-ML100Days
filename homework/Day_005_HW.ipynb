```python
pd.crosstab(app_train["CODE_GENDER"],app_train["TARGET"],margins=True)
```

```python
def percConvert(ser):
  return ser/float(ser[-1])

pect_gender = pd.crosstab(app_train["CODE_GENDER"],app_train["TARGET"],margins=True).apply(percConvert, axis=1)
```

```python
pect_gender
```

```python
pect_gender[0][['F','M','All']]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

N = pect_gender.index.size - 1

paidData = pect_gender[0][['F','M','All']]
notpaidData = pect_gender[1][['F','M','All']]

ind = np.arange(N)    # the x locations for the groups

p2 = ax.bar(ind, notpaidData,  color='b')
p1 = ax.bar(ind, paidData,  color='r')


ax.set_ylabel('Payment Difficulties Percentage')
ax.set_xlabel('Gender')
ax.set_title('Payment Difficulties by gender')

```

```python
N = 3
paidData = pect_gender[0][['F','M','All']]
notpaidData = pect_gender[1][['F','M','All']]

ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, paidData, width)
p2 = plt.bar(ind, notpaidData, width, bottom=paidData)

plt.ylabel('Payment Percentage')
plt.title('Payment Percentage by Gender')
plt.xticks(ind, ('F', 'M', 'All'))
plt.yticks(np.arange(0, 1, 5))
plt.legend((p1[0], p2[0]), ('Female', 'Male'))

plt.show()
```
